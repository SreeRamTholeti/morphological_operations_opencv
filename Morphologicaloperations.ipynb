```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt 
```

```python
#Reading Image and Resizing image
img = cv.imread('pcvimg.jpg',-1) 
dim = (449,403)
img_1 = cv.resize(img, dim, interpolation = cv.INTER_AREA)
```

```python
kernel = np.ones((5,5),np.uint8) 
erosion = cv.erode(img_1,kernel,iterations = 1)
cv.imshow("After_Applying_Erossion",np.hstack((img_1, erosion)))
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
hist = cv.calcHist([erosion],[0],None,[256],[0,256]) 
plt.plot(hist) 
plt.show() 
```

```python
dilation = cv.dilate(img_1,kernel,iterations = 1) 
cv.imshow("After_Applying_Dialation",np.hstack((img_1,dilation)))
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
hist = cv.calcHist([dilation],[0],None,[256],[0,256])
plt.plot(hist) 
plt.show() 
```

```python
opening = cv.morphologyEx(img_1, cv.MORPH_OPEN, kernel) 
cv.imshow("After_Applying_Opening",np.hstack((img_1, opening)))
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
hist = cv.calcHist([opening],[0],None,[256],[0,256]) 
plt.plot(hist) 
plt.show() 
```

```python
closing = cv.morphologyEx(img_1, cv.MORPH_CLOSE, kernel) 
cv.imshow("After_Applying_Closing",np.hstack((img_1,closing)))
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
hist = cv.calcHist([closing],[0],None,[256],[0,256]) 
plt.plot(hist) 
plt.show() 
```

```python
gradient = cv.morphologyEx(img_1, cv.MORPH_GRADIENT, kernel) 
cv.imshow("After_Applying_Gradient",np.hstack((img_1,gradient)))
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
hist = cv.calcHist([gradient],[0],None,[256],[0,256]) 
plt.plot(hist) 
plt.show() 
```

```python
tophat = cv.morphologyEx(img_1, cv.MORPH_TOPHAT, kernel) 
cv.imshow("After_Applying_Tophat",np.hstack((img_1,tophat)))
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
hist = cv.calcHist([tophat],[0],None,[256],[0,256]) 
plt.plot(hist) 
plt.show() 
```

```python
blackhat = cv.morphologyEx(img_1, cv.MORPH_BLACKHAT, kernel) 
cv.imshow("After_Applying_Blackhat",np.hstack((img_1,blackhat)))
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
hist = cv.calcHist([blackhat],[0],None,[256],[0,256]) 
plt.plot(hist) 
plt.show() 
```
